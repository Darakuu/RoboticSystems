```python
import sys
sys.path.append("../")
import time

from lib.data.dataplot import *
from lib.utils.time import *

class S1:
    def __init__(self):
        self.y = 0
        
    def evaluate(self, delta_t, _input):
        y = self.y - (0.5 * delta_t * self.y) + (delta_t * _input)
        self.y = y
        return y

class S2:
    def __init__(self):
        self.y = 0
    
    def evaluate(self, delta_t, _input):
        y = self.y - (3 * delta_t * self.y) + (delta_t * _input)
        self.y = y
        return y

class S3:
    def __init__(self):
        self.y = 0
        
    def evaluate(self, delta_t, _input):
        y = self.y - (-1 * delta_t * self.y) + (4 * delta_t * _input)
        self.y = y
        return y

class CompoundSystem:
    def __init__(self):
        self.s1 = S1()
        self.s2 = S2()
        self.s3 = S3()

    """
    u --- S1 --- S3 --- +() --- y
        |                | + 
        |-S2 ------------|
    """
    
    # NB: S3 dipende da S1!
    def evaluate(self, delta_t, _input):
        out_final = 0.0
        out_s1 = self.s1.evaluate(delta_t, _input)
        out_s2 = self.s2.evaluate(delta_t, _input)
        out_s3 = self.s3.evaluate(delta_t, out_s1)
        out_final = out_s2 + out_s3
        return out_final

dp = DataPlotter()
dp.set_x("time (s)")
dp.add_y("out", "Output")
CompSys = CompoundSystem()

u = 1

t = Time(use_fake_time = True)
t.start()
while t.get() < 10:
    t.sleep(0.01)
    delta_t = t.elapsed()

    y = CompSys.evaluate(delta_t, u)
    
    dp.append_x(t.get())
    dp.append_y("out", y)
dp.plot()
```
